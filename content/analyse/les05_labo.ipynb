## Oefening 1: Numerieke Afgeleiden


### Opdracht 1a

Implementeer een functie `numerical_derivative(f, x, h=1e-5)` die de afgeleide van f op punt x numeriek benadert met de forward difference methode: (f(x+h) - f(x)) / h.

```python
# Jouw code voor opdracht 1a:

# Voorbeeld: f(x) = x²
def f(x):
    return x ** 2

# Numerieke afgeleide
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

# Test op x = 3
x = 3
print(f"f(x) = x² op x = {x}")
print(f"f({x}) = {f(x)}")
print()

# Numerieke benadering met verschillende h
print("Numerieke afgeleide met verschillende stapgroottes:")
for h in [1, 0.1, 0.01, 0.001, 0.0001, 0.00001]:
    deriv = numerical_derivative(f, x, h)
    print(f"  h = {h:8.5f}: f'({x}) ≈ {deriv:.6f}")

print()
print(f"Analytische afgeleide: f'(x) = 2x, dus f'({x}) = {2*x}")
```

### Opdracht 2a

Bereken de afgeleide analytisch (met de hand) voor de volgende functies:

1. f(x) = 5x⁴
2. g(x) = x³ - 2x² + 4x - 1
3. h(x) = 3/x (hint: schrijf als 3x⁻¹)
4. k(x) = √x (hint: schrijf als x^(1/2))
5. m(x) = 2eˣ + 3x²

Implementeer dan zowel de originele functies als hun afgeleiden in Python en verifieer met numerieke afgeleiden.

```python
# Jouw code voor opdracht 2a:


# Originele functies
def f(x): return 5*x**4
def g(x): return x**3 - 2*x**2 + 4*x - 1
def h(x): return 3/x
def k(x): return np.sqrt(x)
def m(x): return 2*np.exp(x) + 3*x**2

# Analytische afgeleiden
def f_prime(x): return 20*x**3
def g_prime(x): return 3*x**2 - 4*x + 4
def h_prime(x): return -3/x**2
def k_prime(x): return 1/(2*np.sqrt(x))
def m_prime(x): return 2*np.exp(x) + 6*x

# Numerieke afgeleide (centrale differentie)
def numerical_derivative(func, x, h=1e-6):
    return (func(x + h) - func(x - h)) / (2*h)

# Testpunten
xs = [0.5, 1, 2]

for x in xs:
    print(f"x = {x}")
    print("f origineel:", f(x), "f' analytisch:", f_prime(x), " numeriek:", numerical_derivative(f, x))
    print("g origineel:", g(x), "g' analytisch:", g_prime(x), " numeriek:", numerical_derivative(g, x))
    print("h origineel:", h(x), "h' analytisch:", h_prime(x), " numeriek:", numerical_derivative(h, x))
    print("k origineel:", k(x), "k' analytisch:", k_prime(x), " numeriek:", numerical_derivative(k, x))
    print("m origineel:", m(x), "m' analytisch:", m_prime(x), " numeriek:", numerical_derivative(m, x))
    print()
```

Implementeer de functies en hun afgeleiden, en verifieer numeriek.

```python
# Jouw code voor opdracht 3a:

# Originele functies
def f(x): return (2*x + 1)**3
def g(x): return np.exp(x**2)
def h(x): return np.cos(3*x)
def k(x): return np.sqrt(x**2 + 1)


# Analytische afgeleiden
def f_prime(x): return 6*(2*x + 1)**2
def g_prime(x): return 2*x*np.exp(x**2)
def h_prime(x): return 3*np.sin(3*x)
def k_prime(x): return x/(2*np.sqrt(x**2 + 1))


# Numerieke afgeleide (centrale differentie)
def numerical_derivative(func, x, h=1e-6):
    return (func(x + h) - func(x - h)) / (2*h)

# Testpunten
xs = [-1.0, 0.0, 0.5, 1.0]

for x in xs:
    print(f"x = {x}")
    print("f origineel:", f(x), "f' analytisch:", f_prime(x), " numeriek:", numerical_derivative(f, x))
    print("g origineel:", g(x), "g' analytisch:", g_prime(x), " numeriek:", numerical_derivative(g, x))
    print("h origineel:", h(x), "h' analytisch:", h_prime(x), " numeriek:", numerical_derivative(h, x))
    print("k origineel:", k(x), "k' analytisch:", k_prime(x), " numeriek:", numerical_derivative(k, x))
    print("m origineel:", m(x), "m' analytisch:", m_prime(x), " numeriek:", numerical_derivative(m, x))
    print()
```
